```python
import requests

prompt = """
News:
Reliance gains after strong earnings.
Banking stocks rally.

Is market bullish or bearish?
"""

response = requests.post(
    "http://localhost:11434/api/generate",
    json={
        "model": "phi3",
        "prompt": prompt,
        "stream": False
    }
)

print(response.json()["response"])
```

```python
import duckdb
from pathlib import Path

# Notebook cwd is often `notebooks/` — use project root so you hit the same DB as `scripts/Get_instruments.py`.
_root = Path.cwd()
if _root.name == "notebooks":
    _root = _root.parent
db_path = _root / "market_data.duckdb"

con = duckdb.connect(str(db_path))

# If this errors, run from project root: `python scripts/Get_instruments.py` (creates + fills `index_constituents`).
df = con.execute("""
SELECT *
FROM index_constituents
WHERE index_name = 'NIFTY_50'
""").fetchdf()

print(df.head())
con.close()
```
